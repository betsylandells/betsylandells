```python
fig, (ax1, ax2) = plt.subplots(2,figsize=(10,10))
fig.suptitle('Energy histograms for shield vs copper object')
ax1.hist(lead_shield,color='gray',label='lead shield')
ax2.hist(lead_object,color='orange',label='copper sphere')
ax1.title.set_text('Lead shield')

ax2.title.set_text('Copper object')
ax2.set_xlabel('Energy MeV')
plt.show()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2,figsize=(10,10))
fig.suptitle('Energy histograms for shield vs copper object')
ax1.hist(copper_shield,color='gray',label='copper shield')
ax2.hist(copper_object,color='orange',label='copper sphere')
ax1.title.set_text('Copper shield')

ax2.title.set_text('Copper object')
ax2.set_xlabel('Energy MeV')
plt.show()
```

```python
ratio_lead = lead_object/lead_shield
```

```python
mean_ratio_pb = np.mean(ratio_lead)
```

```python
ratio_copper = copper_object/copper_shield
```

```python
mean_ratio_cu = np.mean(ratio_copper)
```

The density of lead is $11.29g/cm^{3}$ and the density of copper is $8.96 g/cm^3$. Since the density of copper is much lower, its shield is much less efficient and so the object recieves much more energy, since less has been absorbed. This means that the expected ratio of object:shield energies, the copper shield will be higher. The object:shield ratio for copper shield is 1.577 whereas the ratio for the lead shield is much lower at 0.34, as the shield absorbs much more of the energy from the incoming particles before they reach the object. 

Another note to add is that in the second run, with the copper shield and the copper object, they are the same material, and so are equally absorbant, but with the shield absorbing the particles first. The ratio is 1.57 as once the particles loose some energy in the shield which is only 2cm thick, they go into the object which is a sphere of 30cm radius which therefore absorbs lots of particles with higher energies due to particles created and deflected after hitting the shield.
